```python
import csv
import matplotlib.pyplot as plt
from tabulate import tabulate
```

```python
# Reading the file row by row
# Saving each row in a different list
# 'folder' is a list of lists
def readCSV(filePath):
    
    folder = []
    with open(filePath, newline = '') as file:
        reader = csv.reader(file)
        for row in reader:
            dataset = [float(value) for value in row if value != '']
            folder.append(dataset)
    return folder
```

```python
# File's path
small_filePath = "results/small_full_resources.csv" 

# Reading file
small_dataset = readCSV(small_filePath)
```

```python
colNames = ['m', 'blocks', 'similar_docs', 'total_exe']
print(tabulate(small_dataset, headers = colNames, tablefmt = "grid"))
```

```python
# File's path
medium_filePath = "results/medium_full_resources.csv"

# Reading file
medium_dataset = readCSV(medium_filePath)
```

```python
colNames = ['m', 'blocks', 'similar_docs', 'total_exe']
print(tabulate(medium_dataset, headers = colNames, tablefmt = "grid"))
```

```python
# File's path
big_filePath = "results/big_full_resources.csv"

# Reading file
big_dataset = readCSV(big_filePath)
```

```python
colNames = ['m', 'blocks', 'similar_docs', 'total_exe']
print(tabulate(big_dataset, headers = colNames, tablefmt = "grid"))
```

```python
# given a list, divide it in 3 sublists, return 3 different lists
def dividing_list(data):
    return data[0:4], data[4:8], data[8:12]
```

```python
small_m_64, small_m_128, small_m_256 = dividing_list(small_dataset)
```

```python
medium_m_64, medium_m_128, medium_m_256 = dividing_list(medium_dataset)
```

```python
big_m_64, big_m_128, big_m_256 = dividing_list(big_dataset)
```

```python
# given a list, taking only position 0, 1, 3
def selecting_list(data):
    new_data = []
    for info in data:
        temp = info[0:2]
        temp.append(info[3])
        new_data.append(temp)
    return new_data
```

```python
new_small_m_64 = selecting_list(small_m_64)
new_small_m_128 = selecting_list(small_m_128)
new_small_m_256 = selecting_list(small_m_256)
small = [new_small_m_64, new_small_m_128, new_small_m_256]
```

```python
new_medium_m_64 = selecting_list(medium_m_64)
new_medium_m_128 = selecting_list(medium_m_128)
new_medium_m_256 = selecting_list(medium_m_256)
medium = [new_medium_m_64, new_medium_m_128, new_medium_m_256]
```

```python
new_big_m_64 = selecting_list(big_m_64)
new_big_m_128 = selecting_list(big_m_128)
new_big_m_256 = selecting_list(big_m_256)
big = [new_big_m_64, new_big_m_128, new_big_m_256]
```

```python
def picking_info(data):
    blocks = [entry[1] for entry in data]  
    times = [entry[2] for entry in data]  
    return blocks, times
```

```python
def plot_execution_time(data1, data2, data3):
    
    fig, axs = plt.subplots(1, 3, figsize=(15, 5))
    
    blocks_64_small, times_64_small = picking_info(data1[0])
    blocks_128_small, times_128_small = picking_info(data1[1])
    blocks_256_small, times_256_small = picking_info(data1[2])
    
    blocks_64_medium, times_64_medium = picking_info(data2[0])
    blocks_128_medium, times_128_medium = picking_info(data2[1])
    blocks_256_medium, times_256_medium = picking_info(data2[2])
    
    blocks_64_big, times_64_big = picking_info(data3[0])
    blocks_128_big, times_128_big = picking_info(data3[1])
    blocks_256_big, times_256_big = picking_info(data3[2])
    
    axs[0].plot(blocks_64_small, times_64_small, marker='o', label='SMALL 64 bits')
    axs[0].plot(blocks_128_small, times_128_small, marker='o', label='SMALL 128 bits')
    axs[0].plot(blocks_256_small, times_256_small, marker='o', label='SMALL 256 bits')
    axs[0].set_xlabel('Block length')
    axs[0].set_ylabel('Total execution time')
    axs[0].set_title('Trend of the execution time based on the number of blocks')
    axs[0].legend()
    axs[0].grid(True)
    
    axs[1].plot(blocks_64_medium, times_64_medium, marker='^', label='MEDIUM 64 bits')
    axs[1].plot(blocks_128_medium, times_128_medium, marker='^', label='MEDIUM 128 bits')
    axs[1].plot(blocks_256_medium, times_256_medium, marker='^', label='MEDIUM 256 bits')
    axs[1].set_xlabel('Block length')
    axs[1].set_ylabel('Total execution time')
    axs[1].set_title('Trend of the execution time based on the number of blocks')
    axs[1].legend()
    axs[1].grid(True)
    
    axs[2].plot(blocks_64_big, times_64_big, marker='*', label='BIG 64 bits')
    axs[2].plot(blocks_128_big, times_128_big, marker='*', label='BIG 128 bits')
    axs[2].plot(blocks_256_big, times_256_big, marker='*', label='BIG 256 bits')
    axs[2].set_xlabel('Block length')
    axs[2].set_ylabel('Total execution time')
    axs[2].set_title('Trend of the execution time based on the number of blocks')
    axs[2].legend()
    axs[2].grid(True)
    
    
    plt.tight_layout()
    plt.show()

plot_execution_time(small, medium, big)
```

```python
def extracting_docs(data):
    docs = []
    print(data[3][2])
    docs.append(data[3][2])
    print(data[7][2])
    docs.append(data[7][2])
    print(data[11][2])
    docs.append(data[11][2])
    return docs
```

```python
def plot_bar(values):
    
    colors = ['blue', 'green', 'purple']
    labels = ['m = 64', 'm = 128', 'm = 256']
    plt.bar(labels, values, color=colors)

    plt.title('Number of similar documents according to the length of the signature')
    plt.xlabel('Length of the signature')
    plt.ylabel('Number of similar documents')

    plt.show()
```

```python
values_small = extracting_docs(small_dataset)
plot_bar(values_small)
```

```python
values_medium = extracting_docs(medium_dataset)
plot_bar(values_medium)
```

```python
values_big = extracting_docs(big_dataset)
plot_bar(values_big)
```
